# Tasks 1: Data Cleaning and Preprocessing

#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataset
df = pd.read_csv(r"D:\SaiKet Systems Internship\Telco_Customer_Churn_Dataset .csv")
```

2) Convert categorical variables into numerical representations techniques like one-hot encoding.

```python
# Converting Yes/No columns into binary numeric form
yes_no_cols = ['Partner', 'Dependents', 'PhoneService', 
               'PaperlessBilling', 'Churn']

# Clean up and standardize all yes/no columns before mapping
for col in yes_no_cols:
    df[col] = df[col].astype(str).str.strip().str.title()  # Removes spaces, fixes case
    df[col] = df[col].map({'Yes': 1, 'No': 0})
df[col]
```

```python
#As per the data type TotalCharges it's an object, so we converting into numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'].replace(" ", pd.NA))

# converting null into median values
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True) 

#count of missing values
print(df['TotalCharges'].isna().sum())
```

# Tasks 3: Customer Segmentation

1) Segment customers based on tenure,monthly charges, and contract type.

```python
# Churn rate by Tenure group
df['TenureGroup'] = pd.cut(df['tenure'],bins=[0, 12, 24, 48, 72], labels=['0-1 yr', '1-2 yr', '2-4 yr', '4-6 yr'])

# Churn rate by Charge group
df['ChargeGroup'] = pd.cut(df['MonthlyCharges'],bins=[0, 35, 70, 105], labels=['Low', 'Medium', 'High'])
```

```python
# Churn rate by Tenure group
tenure_churn = df.groupby('TenureGroup',observed=True)['Churn'].value_counts(normalize=True).unstack() * 100
print("\n",tenure_churn)
print("\n" + "-"*30 + "\n")

charge_churn = df.groupby('ChargeGroup',observed=True)['Churn'].value_counts(normalize=True).unstack() * 100
print(charge_churn)
print("\n" + "-"*30 + "\n")

contract_churn = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack() * 100
print(contract_churn)
```

2) Analyze churn rates across these segments

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.countplot(data=df, x='TenureGroup', hue='Churn')
plt.title('Churn by Tenure Group')

plt.subplot(1, 3, 2)
sns.countplot(data=df, x='ChargeGroup', hue='Churn')
plt.title('Churn by Monthly Charges')

plt.subplot(1, 3, 3)
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')

plt.tight_layout()
plt.show()
```

# Tasks 5: Customer Retention Strategies

1) Propose data-driven strategies to reduce churn. 

#Insights 
1) Short tenure customers churn more.
2) Month-to-month contracts have the highest churn.
3) Electronic check payments are linked with higher churn.
4) Customers without tech support / online security churn more.

2) Identify key factors influencing churn and suggest actions.

```python
# Convert 'Churn' to numeric for analysis
df['ChurnFlag'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Correlation with numeric columns
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
corr = df[numeric_cols + ['ChurnFlag']].corr()['ChurnFlag'].sort_values(ascending=False)        
print(corr)
```

| Factor                         | Relationship                    | Suggested Action                                    |
| ------------------------------ | ------------------------------- | --------------------------------------------------- |
| **Tenure**                     | Lower tenure â†’ higher churn     | Engage new users early, provide onboarding benefits |
| **MonthlyCharges**             | Higher charges â†’ higher churn   | Offer personalized or discounted packages           |
| **Contract Type**              | Month-to-month â†’ higher churn   | Encourage switching to longer contracts             |
| **TechSupport/OnlineSecurity** | Absence â†’ higher churn          | Add free trials or bundled offers                   |
| **PaymentMethod**              | Electronic check â†’ higher churn | Promote AutoPay options                             |


3) Calculate the lifetime value (LTV) of customers.


```python
#LTV = MonthlyCharges Ã— Tenure
df['LTV'] = df['MonthlyCharges'] * df['tenure']
df[['customerID', 'MonthlyCharges', 'tenure', 'LTV']].head()
```

```python
print("Average LTV (All Customers):", df['LTV'].mean())
print(df.groupby('Churn')['LTV'].mean())
```

4) Identify High-Value Customers at Risk of Churning

```python
high_value_cust = df['LTV'].quantile(0.75)
high_value_churn = df[(df['LTV'] >= high_value_cust) & (df['Churn'] == 'Yes')]

print("High-value customers at risk of churn:", len(high_value_churn))
high_value_churn[['customerID', 'tenure', 'MonthlyCharges', 'LTV']]
```
